```python
age_as_int = 30 
print(f"age as int - {age_as_int} type = {type(age_as_int)}")
```

```python
age_as_float = 30.0
print(f"age as float - {age_as_float} type = {type(age_as_float)}")
```

```python
temp1 = age_as_int
```

```python
#id(object_name) is a built in function that represents the identity associated with the object 

print(f"ID of age_as_int = {id(age_as_int)}")
print(f"ID of temp1 = {id(temp1)}")
print(f"ID of age_as_float = {id(age_as_float)}")
```

```python
temperature = input("enter the temperature in Farenhite : ")
print(f"Entered temperature value is : {temperature} F and its type is {type(temperature)}")
```

```python
print(f"ID of temperature = {id(temperature)}")
```

```python
temperature = float(temperature) 
print(f"Entered temperature value is : {temperature} F and its type is {type(temperature)}")
print(f"ID of temperature = {id(temperature)}")
```

```python
temperature = temperature * 10.0
print(f"Entered temperature value is : {temperature} F and its type is {type(temperature)}")
print(f"ID of temperature = {id(temperature)}")
```

```python
character = 'V'
ascii_value = ord(character)
print(f"The ASCII value of '{character}' is {ascii_value}")

character_2 = 'P'
ascii_value_2 = ord(character_2)
print(f"The ASCII value of '{character_2}' is {ascii_value_2}")
```
