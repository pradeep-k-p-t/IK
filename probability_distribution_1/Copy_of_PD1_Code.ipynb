## 1. Bernoulli Distribution
### Scenario: Will They Show Up?
Every day, countless patients schedule appointments. But will they show up? Using the Bernoulli Distribution, we can model the probability of patients attending their appointments. This helps in reducing wasted time slots and improving scheduling efficiency.


```python
# Import the required libraries and modules
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, poisson, randint
```

```python
# Case 1: Bernoulli Distribution
# Scenario: Patient appointment no-show rates (show=1, no-show=0)
p_bernoulli = 0.8  # Probability of showing up
bernoulli_trials = bernoulli.rvs(p_bernoulli, size=1000) ## Generate 1000 random samples of patient show-up (1) or no-show (0) with a 80% show-up probability.
```

```python
# Plot Bernoulli distribution
plt.figure(figsize=(8, 6))
plt.hist(bernoulli_trials, bins=2, density=True, alpha=0.6, color='g', edgecolor='black')
plt.title('Bernoulli Distribution (p=0.8)')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.xticks([0, 1])
plt.show()
```

The graph confirms that, according to the Bernoulli distribution with
`p=0.8`, most patients (80%) are expected to show up for their appointments, while a smaller proportion (20%) are likely to miss their appointments.
This visualization helps in understanding the distribution of patient attendance and can assist in resource planning and scheduling in the hospital.

```python
# Additional Analysis and Summary
print(f"Bernoulli Distribution: p={p_bernoulli}")
print(f"Mean of Bernoulli trials: {np.mean(bernoulli_trials)}")
```

The observed mean is very close to the theoretical probability of 0.8, confirming that the Bernoulli trials are correctly modeling the scenario where there is an 80% chance of a patient showing up.
This slight deviation is normal and expected in any empirical experiment, especially with a finite number of trials.

## 2. Binomial Distribution
### Scenario: Tracking the number of successful surgeries out of a set number of operations.
The hospital aims for the highest success rate in surgeries. By applying the Binomial Distribution, we can analyze the probability of successful outcomes over a series of operations. This insight helps in identifying factors that contribute to successful surgeries and areas needing improvement.

```python
# Case 2: Binomial Distribution
# Scenario: Number of successful surgeries out of 20 operations
n_binom = 20  # Number of trials (operations)
p_binom = 0.95  # Probability of success (successful surgery)
binom_trials = binom.rvs(n_binom, p_binom, size=1000) # Generate 1000 random samples representing the number of successful surgeries out of n_binom operations with a success probability of p_binom.
```

```python
# Plot Binomial distribution
plt.figure(figsize=(8, 6))
plt.hist(binom_trials, bins=n_binom+1, density=True, alpha=0.6, color='b', edgecolor='black')
plt.title(f'Binomial Distribution (n={n_binom}, p={p_binom})')
plt.xlabel('Number of Successful Surgeries')
plt.ylabel('Probability')
plt.show()
```

The binomial distribution graph demonstrates that with a high probability of success (95%), most of the outcomes cluster around the higher end of the possible range (19 or 20 successful surgeries out of 20).

This visualization helps in understanding the reliability and consistency of the surgical procedures, confirming that the majority of operations are expected to be successful with only a small chance of fewer successful outcomes.




```python
# Additional Analysis and Summary
print(f"\nBinomial Distribution: n={n_binom}, p={p_binom}")
print(f"Mean of Binomial trials: {np.mean(binom_trials)}")
```

The observed mean is nearly equal to the theoretical mean of 19, confirming that the Binomial trials accurately reflect the expected outcome given a 95% success rate.

This demonstrates the reliability of the surgical procedures, as the average number of successful surgeries closely matches the predicted value.

## 3. Poisson Distribution
### Scenario: Modeling the frequency of emergency room visits per hour.
The emergency room (ER) never rests. Understanding the Poisson Distribution helps us predict the number of ER visits in any given hour. This prediction is crucial for staffing, ensuring that there are always enough doctors and nurses to handle the influx of patients efficiently.

```python
# Case 3: Poisson Distribution
# Scenario: Number of emergency room visits per hour
lambda_poisson = 10  # Average rate of visits per hour
poisson_trials = poisson.rvs(lambda_poisson, size=1000) # Generate 1000 random samples representing the number of emergency room visits per hour with an average rate of lambda_poisson.
```

```python
# Plot Poisson distribution
plt.figure(figsize=(8, 6))
plt.hist(poisson_trials, bins=max(poisson_trials)-min(poisson_trials), density=True, alpha=0.6, color='r', edgecolor='black')
plt.title(f'Poisson Distribution (λ={lambda_poisson})')
plt.xlabel('Number of Visits')
plt.ylabel('Probability')
plt.show()
```

The Poisson distribution graph demonstrates that the number of emergency room visits per hour typically centers around 10, which is the average rate (λ).

Most of the time, the hospital can expect around 10 visits per hour, with a significant probability of seeing a few visits more or fewer.

This information is valuable for hospital resource planning, ensuring that staffing levels are appropriate to handle the expected number of ER visits efficiently.

The graph also shows the range of variability in the number of visits, helping the hospital prepare for slightly busier or quieter hours.

```python
# Additional Analysis and Summary
print(f"\nPoisson Distribution: λ={lambda_poisson}")
print(f"Mean of Poisson trials: {np.mean(poisson_trials)}")
```

The observed mean is very close to the theoretical mean of 10, confirming that the Poisson trials accurately reflect the expected average number of emergency room visits per hour.

In practical terms, the hospital can expect approximately 10 emergency room visits per hour, which aligns with the predicted rate and helps in planning resources and staff allocation effectively.

## 4.Visualizing the Density Functions of the Discrete Distributions

```python
# Parameters for each distribution
p_ber = 0.3  # Probability of success for Bernoulli and Binomial
n_bin = 7    # Number of trials for Binomial
λ_poisson = 10  # Average rate for Poisson


# Generate data points for each distribution
# Bernoulli distribution (for illustration)
data_ber = bernoulli.rvs(p_ber, size=1000)

# Binomial distribution
data_bin = binom.rvs(n_bin, p_ber, size=1000)

# Poisson distribution
data_poisson = poisson.rvs(λ_poisson, size=1000)


# Plotting
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Bernoulli distribution plot
axs[0, 0].hist(data_ber, bins=[-0.5, 0.5, 1.5], density=True, alpha=0.75)
axs[0, 0].set_title('Bernoulli Distribution (p=0.3)')
axs[0, 0].set_xticks([0, 1])
axs[0, 0].set_xticklabels(['Failure', 'Success'])
axs[0, 0].set_ylabel('Probability Density')

# Binomial distribution plot
axs[0, 1].hist(data_bin, bins=np.arange(0, n_bin + 2) - 0.5, density=True, alpha=0.75)
axs[0, 1].set_title(f'Binomial Distribution (n={n_bin}, p=0.3)')
axs[0, 1].set_xlabel('Number of Successes')
axs[0, 1].set_ylabel('Probability Density')

# Poisson distribution plot
axs[1, 0].hist(data_poisson, bins=np.arange(0, 30 + 2) - 0.5, density=True, alpha=0.75)
axs[1, 0].set_title(f'Poisson Distribution (λ={λ_poisson})')
axs[1, 0].set_xlabel('Number of Events')
axs[1, 0].set_ylabel('Probability Density')



plt.tight_layout()
plt.show()
```

**Bernoulli Distribution:** Confirms the expected probabilities for a binary
outcome (30% show-up, 70% no-show).

**Binomial Distribution:** Shows the probability distribution for the number of successes out of a fixed number of trials, with a peak around the expected number of successes given the probability.

**Poisson Distribution:** Demonstrates the expected number of events per interval (10 visits per hour) with variability, highlighting the most likely number of visits and the spread around this average.
